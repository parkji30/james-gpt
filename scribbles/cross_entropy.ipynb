```python
import torch

def cross_entropy(y, y_pred):
    """cross entropy loss is used to measure how far off your models probabilities are from the truth values."""
    return - (y * torch.log(y_pred) + (1- y) * torch.log( 1 - y_pred))


y_target = torch.Tensor([0, 1, 0])
y_pred = torch.Tensor([0.1, 0.8, 0.1])
cross_entropy(y_target, y_pred)
```
