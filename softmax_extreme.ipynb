```python
import torch
import torch.nn.functional as F
```

```python
def softmax(x):
    """This operators on a sequence of data"""
    exps = torch.exp(x)
    return exps / exps.sum()

x = torch.tensor([
    [2.0, 1.0, 0.0],
    [1.0, 3.0, 2.0],
])

print(torch.sum(softmax(x)))
print(torch.vmap(sum)(torch.vmap(softmax)(x)))
```

```python
#vmap over column

original = torch.vmap(softmax, in_dims=1, out_dims=1)(x)
original
```

```python
# vmap over column and spit out answer along row

reverse = torch.vmap(softmax, in_dims=1, out_dims=0)(x)
reverse
```

```python
torch.all(reverse.T == original)
```

```python
# softmax extreme example

base = torch.tensor([1.0, 2.0, 3.0])

for scale in [1, 2, 5, 10]:
    logits = scale * base


    print(logits)
    print(softmax(logits))
```
