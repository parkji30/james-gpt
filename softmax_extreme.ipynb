```python
import torch
import torch.nn.functional as F
```

```python
def softmax(x):
    """This operators on a sequence of data"""
    exps = torch.exp(x)
    return exps / exps.sum()

x = torch.tensor([
    [2.0, 1.0, 0.0],
    [1.0, 3.0, 2.0],
])

print(torch.sum(softmax(x)))
print(torch.vmap(sum)(torch.vmap(softmax)(x)))
```

```python
#vmap over column

original = print(torch.vmap(softmax, in_dims=1, out_dims=1)(x))
```

```python
# vmap over column and spit out answer along row

reverse = print(torch.vmap(softmax, in_dims=1, out_dims=0)(x))
```

```python
reverse.T == original
```
